```python
import pandas as pd
import sqlite3
```

# Run SQL with help of pandas

```python
# Load the CSV into a DataFrame
df = pd.read_csv('used_car_price.csv')

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

# Write the DataFrame to the SQLite database as a table
df.to_sql('my_table', conn, index=False, if_exists='replace')
```

```python
df.head()
```

```python
# Run an SQL query (e.g., select specific columns and filter rows)
query = "SELECT * FROM my_table;"

# Display the results
result_df = pd.read_sql_query(query, conn)
result_df.head()
```

```python
#Brand name "Genesis"
query = "SELECT * from my_table where brand=='Mercedes-Benz';"

#Display the result
result_df = pd.read_sql_query(query, conn)
result_df.head()
```
